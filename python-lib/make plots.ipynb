# Iris Data - TSNE

Using pyplot to interact with 2d and 3d TSNE transforms of iris dataset.

```python
from plotly import express as px
import pandas as pd
from sklearn import datasets
from sklearn.manifold import TSNE
import plotly.graph_objs as go
import plotly.io as pio
```

```python
data = datasets.load_iris()
X = data.data
y = data.target
df = pd.DataFrame(X)
```

```python
df
```

```python
tsne_model_2d = TSNE(2)
data_2d = pd.DataFrame(tsne_model_2d.fit_transform(X))
tsne_model_3d = TSNE(3)
data_3d = pd.DataFrame(tsne_model_3d.fit_transform(X))
```

```python
data_2d
```

```python
plot_2d = px.scatter(data_2d, 0, 1, color=y)
plot_2d.show()
```

```python
plot_3d = px.scatter_3d(data_3d, 0, 1, 2, color=y)
plot_3d.show()
```

```python
pio.write_json(plot_3d, '../plots/plot_3d.json')
pio.write_json(plot_2d, '../plots/plot_2d.json')
```
